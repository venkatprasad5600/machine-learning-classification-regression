```python
import pandas as pd 
```

```python
# laoding dataset 
df=pd.read_csv('C:/Users/Prasad/Downloads/student_perfomance.csv')
df
```

```python
# cleaning the data 
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# feature engineering 
df['Hours_Attendance']=df['Hours_Studied']*df['Attendance (%)']
df
```

```python
df.columns
```

```python
# feature scalling 
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
features = ['Hours_Studied', 'Attendance (%)', 'Previous_Marks', 'Study_Environment (0=Poor,1=Average,2=Good)', 'Hours_Attendance']
X = df[features]
X_scaled=scalar.fit_transform(X)
y=df['Exam_Score']
```

```python
# spilting the data 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,train_size=0.2,random_state=42)
```

```python
# EDA 
import matplotlib.pyplot as plt 
plt.scatter(x='Hours_Studied',y='Exam_Score',data=df)
plt.show()
```

```python
# model building using linear regression
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score


# polynomial reg
poly=PolynomialFeatures(degree=2,include_bias=False)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
# model
lr=LinearRegression()
lr.fit(X_train_poly,y_train)

# prediction
y_pred=lr.predict(X_test_poly)
print('mae',mean_absolute_error(y_test,y_pred))
mse=mean_squared_error(y_test,y_pred)
print('mse',mse)
print(np.sqrt(mse))
print('R2 score',r2_score(y_test,y_pred))


plt.figure(figsize=(8,6))
plt.scatter(y_test,y_pred,color='blue')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()],'r--',linewidth=2)
plt.xlabel('actual exam scores')
plt.ylabel('predicted exam scores')
plt.title('polynomial regression:actual vs predicted')
plt.show()
```
